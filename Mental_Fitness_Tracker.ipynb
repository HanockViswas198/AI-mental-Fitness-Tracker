```python
df=data
df.info()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Create a new DataFrame to store the transformed data
encoded_df = pd.DataFrame()

for column in df.columns:
    if df[column].dtype == 'object':
        # Create a one-hot encoder instance
        encoder = OneHotEncoder(sparse=False, drop='first')

        # Fit and transform the column
        encoded_column = encoder.fit_transform(df[column].values.reshape(-1, 1))

        # Create new column names based on the unique labels in the original column
        column_names = [column + '_' + label for label in encoder.get_feature_names_out([column])]

        # Append the encoded column to the new DataFrame
        encoded_df = pd.concat([encoded_df, pd.DataFrame(encoded_column, columns=column_names)], axis=1)
    else:
        # Append non-object columns directly to the new DataFrame
        encoded_df = pd.concat([encoded_df, df[column]], axis=1)

# Display the transformed DataFrame
print(encoded_df)
```

```python
X= df.drop('mental_fitness',axis=1)
y= df['mental_fitness']

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=.20, random_state=2)
```

```python
print("xtrain: ", xtrain.shape);
print("xtest: ", xtest.shape);
print("\n ytrain: ", ytrain.shape)
print("ytest: ",ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr= LinearRegression()
lr.fit(xtrain,ytrain)

#Model evaluation for Training set
ytrain_pred =  lr.predict(xtrain)


#The mean square method for Average between observed abnd predicted
mse = mean_squared_error(ytrain, ytrain_pred)


#Root Mean Square Error
rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))

#The coefficient of determination

r2 = r2_score(ytrain, ytrain_pred)


print("The Linear Regression model performance for training set")
print("--------------------------------------------")
print('MSE is {}' , format(mse))
print('RMSE is {}', format(rmse))
print('R2 score is {}', format(r2))



```

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor()
rf.fit(xtrain,ytrain)

# Model evaluation for training set
ytrain_pred = rf.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
r2 = r2_score(ytrain, ytrain_pred)




print("The Random Forest Regresser model performance for training set")
print("--------------------------------------------")
print('MSE is {}' , format(mse))
print('RMSE is {}', format(rmse))
print('R2 score is {}', format(r2))


```
